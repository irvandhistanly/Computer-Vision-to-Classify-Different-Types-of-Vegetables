# Data Inference

## Import Libraries and Files

Importing necessary libraries and the files that were saved from our model creation

```python
# Load library
import cv2
import numpy as np
import tensorflow as tf
from IPython.display import Image, display
```

```python
# Load model

model = tf.keras.models.load_model('model_improved.keras')
```

## Creating Relevant Functions

1. To pre-processed images
2. Predicting the image's pixel forms
3. Showing the image
4. Predicting the image's label

```python
# Function to preprocess the input image (same as model training)
def preprocess_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (224, 224))  # Resize the image to match the input size of your model
    img = img / 255.0  # Normalize pixel values to [0, 1]
    return img
```

```python
# Function to perform inference on a single image
def predict_single_image(image_path):
    preprocessed_img = preprocess_image(image_path)
    input_img = np.expand_dims(preprocessed_img, axis=0)  # Add batch dimension
    prediction = model.predict(input_img)
    return prediction
```

```python
def display_image(image_path):
    # Display the image
    display(Image(filename=image_path))
```

```python
class_names = ['Bean', 'Bitter_Gourd', 'Bottle_Gourd', 'Brinjal', 'Broccoli', 'Cabbage', 'Capsicum', 'Carrot', 'Cauliflower', 'Cucumber', 'Papaya', 'Potato', 'Pumpkin', 'Radish', 'Tomato']

# Function to get the predicted class label
def get_predicted_class(prediction):
    predicted_class_index = np.argmax(prediction)
    predicted_class = class_names[predicted_class_index]
    return predicted_class
```

## Predicting

We loaded 2 brand new, unseen vegetables pictures, cucumber and pumpkin. We preprocessed and display the image, predict the pixel form, as well as predict the label.

The model was shown to accurately predict both vegetables

```python
image_path = 'Vegetable Images\Inference\inference_cucumber.jpeg'
display_image(image_path)
```

```python
image_path = 'Vegetable Images\Inference\inference_cucumber.jpeg'
prediction1 = predict_single_image(image_path)
print("Prediction:", prediction1)
```

```python
predicted_class = get_predicted_class(prediction1)
print("Predicted class:", predicted_class)
```

```python
image_path = 'Vegetable Images\Inference\inference_pumpkins.jpg'
display_image(image_path)
```

```python
image_path = 'Vegetable Images\Inference\inference_pumpkins.jpg'
prediction2 = predict_single_image(image_path)
print("Prediction:", prediction2)
```

```python
predicted_class = get_predicted_class(prediction2)
print("Predicted class:", predicted_class)
```
